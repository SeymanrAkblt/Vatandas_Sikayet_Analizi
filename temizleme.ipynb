```python
def temizle(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(rf"[{re.escape(string.punctuation)}]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python

sikayet_kelimeleri = [
    'şikayet', 'rezalet','yapılmadı','istemiyoruz', 'yetersiz','olmasın', 'kötü', 'berbat', 'bozuk','tehlikeli',
    'pis', 'çöp', 'dökülüyor', 'kirli', 'sorun', 'arızalı','yapılmıyor','yapılacak',
    'eksik', 'gürültü', 'çalışmıyor', 'şikayetçiyim', 'şikayetim var','kesiliyor',
    'rahatsız', 'düzenlenmeli', 'memnun değilim', 'arızalandı', 'yavaş','sıkıntılı',
    'su yok', 'ışık yok', 'çukur','çamur','çöplük','deli','şikayeti','yüzünden',
    'yok','güzel değil','perişan','iğrenç','gelmiyor','gitmiyor','fena','yavaş',
    'gelecek','sorunlu','yapın artık','kokuyor','yanmıyor','birikti','sökülmüş','yapılmıyor','yoktu',
    'yapılmadı','çekmiyor','hiçbir zaman','yeter artık','bir türlü','bozuldu','koku','yavaş','gürültü',
]

def anahtar_kelime_etiketle(text):
    text = str(text).lower()
    for kelime in sikayet_kelimeleri:
        if kelime in text:
            return 1
    return 0

df['etiket'] = df['temiz_yorum'].apply(anahtar_kelime_etiketle)
```
